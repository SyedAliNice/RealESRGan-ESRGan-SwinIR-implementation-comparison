```python
import torch
from PIL import Image
import numpy as np
from basicsr.archs.rrdbnet_arch import RRDBNet
from basicsr.utils.download_util import load_file_from_url
from realesrgan import RealESRGANer
import os
import gc

def enhance_image(input_path, output_path, model_name='RealESRGAN_x4plus', tile_size=400):
    """
    Enhance an image using Real-ESRGAN with memory optimization

    Args:
        input_path: Path to input image
        output_path: Path to save enhanced image
        model_name: Name of model to use (default: RealESRGAN_x4plus)
        tile_size: Size of tiles for processing (default: 400)
    """
    # Set PyTorch memory allocator settings
    torch.cuda.empty_cache()
    if torch.cuda.is_available():
        # Try to prevent memory fragmentation
        os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'max_split_size_mb:128'

    # Configure device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    try:
        # Initialize model
        if model_name == 'RealESRGAN_x4plus':
            model = RRDBNet(num_in_ch=3, num_out_ch=3, num_feat=64, num_block=23, num_grow_ch=32, scale=4)
            netscale = 4
            model_path = os.path.join('weights', 'RealESRGAN_x4plus.pth')
            if not os.path.exists(model_path):
                model_path = load_file_from_url(
                    'https://github.com/xinntao/Real-ESRGAN/releases/download/v0.1.0/RealESRGAN_x4plus.pth',
                    model_dir='weights')

        # Initialize upsampler with memory optimization settings
        use_half = device.type == 'cuda'
        upsampler = RealESRGANer(
            scale=netscale,
            model_path=model_path,
            model=model,
            tile=tile_size,  # Use tiling to process large images
            tile_pad=10,
            pre_pad=0,
            half=use_half,
            device=device
        )

        # Read and preprocess image
        img = Image.open(input_path).convert('RGB')

        # Get image dimensions
        width, height = img.size
        print(f"Processing image of size {width}x{height}")

        # Convert to numpy array
        img_np = np.array(img)

        # Process image with automatic memory management
        try:
            output, _ = upsampler.enhance(img_np)

            # Save enhanced image
            enhanced_img = Image.fromarray(output)
            enhanced_img.save(output_path)
            print(f'Enhanced image saved to {output_path}')

        except RuntimeError as e:
            if "out of memory" in str(e):
                # If we still get OOM, try with larger tile size or smaller chunks
                torch.cuda.empty_cache()
                gc.collect()

                # Try with larger tile size
                print(f"Retrying with larger tile size...")
                upsampler.tile = tile_size * 2
                output, _ = upsampler.enhance(img_np)

                enhanced_img = Image.fromarray(output)
                enhanced_img.save(output_path)
                print(f'Enhanced image saved to {output_path} (with adjusted tile size)')
            else:
                raise e

    except Exception as e:
        print(f'Error during enhancement: {str(e)}')
        if device.type == 'cuda':
            print("\nTroubleshooting tips:")
            print("1. Try reducing the tile_size (current:", tile_size, ")")
            print("2. Free up GPU memory by closing other applications")
            print("3. Try processing a smaller version of the image")

            # Print GPU memory stats
            if torch.cuda.is_available():
                print("\nGPU Memory Status:")
                print(f"Total GPU memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
                print(f"Allocated: {torch.cuda.memory_allocated() / 1e9:.2f} GB")
                print(f"Cached: {torch.cuda.memory_reserved() / 1e9:.2f} GB")

    finally:
        # Cleanup
        torch.cuda.empty_cache()
        gc.collect()

def main():
    # Example usage
    input_image = "/content/orb_stitched.jpg"  # Replace with your input image path
    output_image = "/content/enhanced_stitch.jpg"  # Replace with desired output path

    # Create weights directory if it doesn't exist
    os.makedirs('weights', exist_ok=True)

    # Start with a conservative tile size
    enhance_image(input_image, output_image, tile_size=200)

if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from PIL import Image
import numpy as np
import os
import math
import gc
from collections import OrderedDict
from torch.nn import init

class RRDB(nn.Module):
    def __init__(self, nf, gc=32):
        super(RRDB, self).__init__()
        self.RDB1 = ResidualDenseBlock_5C(nf, gc)
        self.RDB2 = ResidualDenseBlock_5C(nf, gc)
        self.RDB3 = ResidualDenseBlock_5C(nf, gc)

    def forward(self, x):
        out = self.RDB1(x)
        out = self.RDB2(out)
        out = self.RDB3(out)
        return out * 0.2 + x

class ResidualDenseBlock_5C(nn.Module):
    def __init__(self, nf=64, gc=32, bias=True):
        super(ResidualDenseBlock_5C, self).__init__()
        self.conv1 = nn.Conv2d(nf, gc, 3, 1, 1, bias=bias)
        self.conv2 = nn.Conv2d(nf + gc, gc, 3, 1, 1, bias=bias)
        self.conv3 = nn.Conv2d(nf + 2 * gc, gc, 3, 1, 1, bias=bias)
        self.conv4 = nn.Conv2d(nf + 3 * gc, gc, 3, 1, 1, bias=bias)
        self.conv5 = nn.Conv2d(nf + 4 * gc, nf, 3, 1, 1, bias=bias)
        self.lrelu = nn.LeakyReLU(negative_slope=0.2, inplace=True)

    def forward(self, x):
        x1 = self.lrelu(self.conv1(x))
        x2 = self.lrelu(self.conv2(torch.cat((x, x1), 1)))
        x3 = self.lrelu(self.conv3(torch.cat((x, x1, x2), 1)))
        x4 = self.lrelu(self.conv4(torch.cat((x, x1, x2, x3), 1)))
        x5 = self.conv5(torch.cat((x, x1, x2, x3, x4), 1))
        return x5 * 0.2 + x

class ESRGAN(nn.Module):
    def __init__(self, in_nc=3, out_nc=3, nf=64, nb=23):
        super(ESRGAN, self).__init__()
        self.conv_first = nn.Conv2d(in_nc, nf, 3, 1, 1, bias=True)
        self.RRDB_trunk = nn.Sequential(*[RRDB(nf) for _ in range(nb)])
        self.trunk_conv = nn.Conv2d(nf, nf, 3, 1, 1, bias=True)

        # Upsampling
        self.upconv1 = nn.Conv2d(nf, nf, 3, 1, 1, bias=True)
        self.upconv2 = nn.Conv2d(nf, nf, 3, 1, 1, bias=True)
        self.HRconv = nn.Conv2d(nf, nf, 3, 1, 1, bias=True)
        self.conv_last = nn.Conv2d(nf, out_nc, 3, 1, 1, bias=True)

        self.lrelu = nn.LeakyReLU(negative_slope=0.2, inplace=True)

    def forward(self, x):
        fea = self.conv_first(x)
        trunk = self.trunk_conv(self.RRDB_trunk(fea))
        fea = fea + trunk

        fea = self.lrelu(self.upconv1(F.interpolate(fea, scale_factor=2, mode='nearest')))
        fea = self.lrelu(self.upconv2(F.interpolate(fea, scale_factor=2, mode='nearest')))
        out = self.conv_last(self.lrelu(self.HRconv(fea)))

        return out

class ESRGANProcessor:
    def __init__(self, model_path, device='cuda'):
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        self.model = ESRGAN().to(self.device)
        self.load_model(model_path)
        self.model.eval()

    def load_model(self, model_path):
        state_dict = torch.load(model_path, map_location=self.device)
        if 'params_ema' in state_dict:
            state_dict = state_dict['params_ema']
        # Remove 'module.' prefix if it exists
        new_state_dict = OrderedDict()
        for k, v in state_dict.items():
            name = k.replace('module.', '')
            new_state_dict[name] = v
        self.model.load_state_dict(new_state_dict)

    def process_tile(self, img, tile_size):
        batch, channel, height, width = img.shape
        output_height = height * 4
        output_width = width * 4
        output = torch.zeros((batch, channel, output_height, output_width), device=self.device)

        tiles_x = math.ceil(width / tile_size)
        tiles_y = math.ceil(height / tile_size)

        for y in range(tiles_y):
            for x in range(tiles_x):
                # Extract tile
                ofs_x = x * tile_size
                ofs_y = y * tile_size

                # Handle border cases
                end_x = min((x + 1) * tile_size, width)
                end_y = min((y + 1) * tile_size, height)

                tile = img[:, :, ofs_y:end_y, ofs_x:end_x]

                # Process tile
                with torch.no_grad():
                    output_tile = self.model(tile)

                # Put tile into output image
                output[:, :, ofs_y*4:end_y*4, ofs_x*4:end_x*4] = output_tile

        return output

def enhance_image(input_path, output_path, model_path, tile_size=100):
    """
    Enhance an image using ESRGAN with memory optimization

    Args:
        input_path: Path to input image
        output_path: Path to save enhanced image
        model_path: Path to ESRGAN model weights
        tile_size: Size of tiles for processing
    """
    try:
        # Clear GPU memory
        torch.cuda.empty_cache()
        gc.collect()

        # Set memory allocator settings
        if torch.cuda.is_available():
            os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'max_split_size_mb:128'

        # Initialize processor
        processor = ESRGANProcessor(model_path)

        # Load and preprocess image
        img = Image.open(input_path).convert('RGB')
        img_np = np.array(img)

        # Convert to tensor
        img_tensor = torch.from_numpy(img_np).float().div(255.)
        img_tensor = img_tensor.permute(2, 0, 1).unsqueeze(0).to(processor.device)

        try:
            # Process image with tiling
            output = processor.process_tile(img_tensor, tile_size)

            # Convert back to image
            output = output.squeeze(0).permute(1, 2, 0).clamp(0, 1)
            output = (output * 255).cpu().numpy().astype(np.uint8)

            # Save result
            Image.fromarray(output).save(output_path)
            print(f'Enhanced image saved to {output_path}')

        except RuntimeError as e:
            if "out of memory" in str(e):
                print("Out of memory. Trying with smaller tile size...")
                torch.cuda.empty_cache()
                gc.collect()
                # Retry with smaller tile size
                output = processor.process_tile(img_tensor, tile_size // 2)
                output = output.squeeze(0).permute(1, 2, 0).clamp(0, 1)
                output = (output * 255).cpu().numpy().astype(np.uint8)
                Image.fromarray(output).save(output_path)
                print(f'Enhanced image saved to {output_path} (with reduced tile size)')
            else:
                raise e

    except Exception as e:
        print(f'Error during enhancement: {str(e)}')
        if torch.cuda.is_available():
            print("\nGPU Memory Status:")
            print(f"Total: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
            print(f"Allocated: {torch.cuda.memory_allocated() / 1e9:.2f} GB")
            print(f"Cached: {torch.cuda.memory_reserved() / 1e9:.2f} GB")

    finally:
        # Cleanup
        torch.cuda.empty_cache()
        gc.collect()

def main():
    input_image = "/content/orb_stitched.jpg"  # Replace with your input image path
    output_image = "/content/enhanced_output1.jpg"  # Replace with desired output path
    model_path = "/content/RRDB_ESRGAN_x4.pth"  # Replace with path to ESRGAN weights

    enhance_image(input_image, output_image, model_path, tile_size=100)

if __name__ == "__main__":
    main()
```
